```python
results_19 = pd.read_csv(f'datasets/2019/league/2019-results.csv')
player_info_19 = pd.read_csv(f'datasets/2019/player-2019-info.csv')

forwardslst_19 = pd.read_csv(f'datasets/2019/playerstats/2019-forwards.csv')
midfielderslst_19 = pd.read_csv(f'datasets/2019/playerstats/2019-midfielders.csv')
defenderslst_19 = pd.read_csv(f'datasets/2019/playerstats/2019-defenders.csv')
keeperslst_19 = pd.read_csv(f'datasets/2019/playerstats/2019-keepers.csv')
forwards90lst_19 = pd.read_csv(f'datasets/2019/playerstats/2019-forwards90.csv')
midfielders90lst_19 = pd.read_csv(f'datasets/2019/playerstats/2019-midfielders90.csv')
defenders90lst_19 = pd.read_csv(f'datasets/2019/playerstats/2019-defenders90.csv')
keepers90lst_19 = pd.read_csv(f'datasets/2019/playerstats/2019-keepers90.csv')
disciplinelst_19 = pd.read_csv(f'datasets/2019/playerstats/2019-discipline.csv')
best_eleven_19 = pd.read_csv(f'datasets/2019/playerstats/2019-best_eleven.csv')
rated_goalscorers_19 = pd.read_csv(f'datasets/2019/playerstats/2019-goalscorers.csv')
rated_assists_19 = pd.read_csv(f'datasets/2019/playerstats/2019-assists.csv')
forwards_line_19 = pd.read_csv(f'datasets/2019/playerstats/2019-forwards-line.csv')
midfielders_line_19 = pd.read_csv(f'datasets/2019/playerstats/2019-midfielders-line.csv')
defenders_line_19 = pd.read_csv(f'datasets/2019/playerstats/2019-defenders-line.csv')
keepers_line_19 = pd.read_csv(f'datasets/2019/playerstats/2019-keepers-line.csv')

regular_season_19 = pd.read_csv(f'datasets/2019/league/2019-regular-standings.csv')
playoffs_19 = pd.read_csv(f'datasets/2019/league/2019-playoffs.csv')
championship_19 = pd.read_csv(f'datasets/2019/league/2019-championship.csv')

team_radar_19 = pd.read_csv(f'datasets/2019/league/2019-radar.csv')
power_19 = pd.read_csv(f'datasets/2019/league/2019-power_rankings.csv')
```

```python
season_totals_19 = pd.concat([regular_season_19,playoffs_19,championship_19]).groupby('team').sum()
season_totals_19.insert(8,'GA',season_totals_19['Goal'] - season_totals_19['gd'])
season_totals_19.insert(2,'PPG',round(season_totals_19['points'] / season_totals_19['matches'],2))
season_totals_19 = season_totals_19.reset_index()
season_totals_19 = season_totals_19.sort_values(by='points',ascending=False)
```
